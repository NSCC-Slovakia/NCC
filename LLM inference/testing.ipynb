```python
df['prediction'] = None

for index, row in df.iterrows():

    pom = row["answer"].lstrip("<pad> The correct option is")
    print(pom)
    if row["opa"] in pom or "a" in pom:
        df.at[index, "prediction"] = 0
    elif row["opb"] in pom or "b" in pom:  
        df.at[index, "prediction"] = 1
    elif row["opc"] in pom or "c" in pom:
        df.at[index, "prediction"] = 2
    elif row["opd"] in pom or "d" in pom:  
        df.at[index, "prediction"] = 3
    else:
        df.at[index, "prediction"] = "invalid"
        print("Invalid detected: ", pom, index)
    #t.at[index, "answer"] = answer
```
